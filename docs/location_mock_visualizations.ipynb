```python
# import relevant librarires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

# load in sample data
df = pd.read_csv('../data/clean_data/sample_location_data.csv')

# what does the data look like?
df.head()
```

```python
tconst_to_filter = 'tt0944947'  # example value

# Filter the DataFrame for the given tconst
df_filtered = df[df['tconst'] == tconst_to_filter]

# make map size smaller for better visualization
m = folium.Map(tiles='CartoDB positron', no_wrap = True, zoom_start=2)

# Add points to the map
for idx, row in df_filtered.iterrows():
    tooltip = folium.Tooltip(row['location'])
    folium.Marker(
        [row['latitude'], row['longitude']],
        tooltip=tooltip
    ).add_to(m)

# Get the map title from the primaryTitle column of the filtered DataFrame
if not df_filtered.empty:
    map_title = df_filtered['primaryTitle'].iloc[0]
    title_html = f'''
                 <h3 align="center" style="font-size:20px"><b>{map_title}</b></h3>
                 '''
    m.get_root().html.add_child(folium.Element(title_html))

m 
```
