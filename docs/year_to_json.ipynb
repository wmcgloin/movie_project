```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('../data/clean_data/all_locations.csv')

df1.head(10)
```

```python
# create a list of unique tconst values
tconst_list = df1['tconst'].unique()

# first 10 tconst values
tconst_list[:10]
```

```python
df2 = pd.read_csv('../data/clean_data/financial_info.csv')

df2.head(10)
```

```python
# remove rows with missing values
df2 = df2.dropna()

# create a list of unique tconst values
tconst_list2 = df2['tconst'].unique()

# first 10 tconst values
tconst_list2[:10]
```

```python
# only keep tconst values that are in both lists
tconst_list = np.intersect1d(tconst_list, tconst_list2)
```

```python
df3 = pd.read_csv('../data/clean_data/movie_level_data.csv')

# filter for only titleType == 'movie'
df3 = df3[df3['titleType'] == 'movie']

# only keep tconst and startYear columns
df3 = df3[['tconst', 'startYear']]

# rename columns
df3.columns = ['tconst', 'Year']

df3.shape
```

```python
# only keep rows with tconst values in tconst_list
df3 = df3[df3['tconst'].isin(tconst_list)]

df3.shape
```

```python
df3.head(20)
```

```python
# convert to json
df3.to_json('../data/clean_data/release_years.json', orient='records')
```
