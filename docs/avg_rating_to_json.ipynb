```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/clean_data/movie_level_data.csv')

df.head()
```

```python
# filter for only titleType == 'movie'
df = df[df['titleType'] == 'movie']

# only keep tconst and averageRating columns
df = df[['tconst', 'averageRating']]

# rename columns
df.columns = ['tconst', 'rating']

df.head(10)
```

```python
# Group by the 'rating' column and count the number of occurrences
rating_counts = df.groupby('rating').size()

print(rating_counts)
```

```python
# convert rating_counts to a DataFrame
rating_counts = rating_counts.reset_index()

# rename columns
rating_counts.columns = ['Rating', 'Number of Movies']

rating_counts.head(20)
```

```python
# pandas to dict
# rating_counts_dict = rating_counts.to_dict(orient='records')
```

```python
# convert to json
rating_counts.to_json('../data/clean_data/rating_counts.json', orient='records')
```

```python
df1 = pd.read_csv('../data/clean_data/all_locations.csv')

df1.head(10)
```

```python
# create a list of unique tconst values
tconst_list = df1['tconst'].unique()

# first 10 tconst values
tconst_list[:10]
```

```python
df2 = pd.read_csv('../data/clean_data/financial_info.csv')

df2.head(10)
```

```python
# remove rows with missing values
df2 = df2.dropna()

# create a list of unique tconst values
tconst_list2 = df2['tconst'].unique()

# first 10 tconst values
tconst_list2[:10]
```

```python
# only keep tconst values that are in both lists
tconst_list = np.intersect1d(tconst_list, tconst_list2)
```

```python
df3 = pd.read_csv('../data/clean_data/movie_level_data.csv')

# filter for only titleType == 'movie'
df3 = df3[df3['titleType'] == 'movie']

# only keep tconst and averageRating columns
df3 = df3[['tconst', 'averageRating']]

# rename columns
df3.columns = ['tconst', 'Rating']

df3.shape
```

```python
# only keep rows with tconst values in tconst_list
df3 = df3[df3['tconst'].isin(tconst_list)]

df3.shape
```

```python
df3.head(20)
```

```python
# convert to json
df3.to_json('../data/clean_data/movie_ratings.json', orient='records')
```
