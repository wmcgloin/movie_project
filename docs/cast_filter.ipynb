```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/clean_data/cast_and_crew.csv')

df.head(10)
```

```python
df[(df['numVotes'] > 500000) & (df['averageRating'] > 4)].drop_duplicates(subset = ['tconst', 'primaryName'])['primaryName'].value_counts().head(20)
```

```python
# only keep the columns primaryName, tconst, averageRating, numVotes
df = df[['primaryName', 'tconst', 'averageRating', 'numVotes']]

df.shape
```

```python
# drop duplicates
df = df.drop_duplicates(subset = ['tconst', 'primaryName'])

df.shape
```

```python
# if a primaryName appears less than 9 times, drop it
df = df.groupby('primaryName').filter(lambda x: len(x) > 9)

df.shape
```

```python
df1 = pd.read_csv('../data/clean_data/movie_level_data.csv')

# only keep rows where df1 titleType is movie
df1 = df1[df1['titleType'] == 'movie']

df1.head()
```

```python
# create a list of all unique tconst values
tconst_list = df1['tconst'].unique()

# only keep rows where df tconst is in tconst_list
df = df[df['tconst'].isin(tconst_list)]

# remove tconst column
df = df.drop(columns = ['tconst'])

df.shape
```
