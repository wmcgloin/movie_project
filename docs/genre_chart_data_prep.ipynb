```python
# find max length of a list in the genre column
df['genres'].apply(lambda x: len(x.split(','))).max()
```

```python
# First ensure all values in 'genres' are strings
df['genres'] = df['genres'].astype(str)

# Remove rows with NaN value or empty string for the 'genres' column after conversion
df = df[df['genres'] != 'nan']

# Find the maximum number of genres in any row
max_genres = df['genres'].str.split(',').map(lambda x: len(x) if isinstance(x, list) else 0).max()

# Create new columns for each genre
genre_cols = [f'genre{i+1}' for i in range(max_genres)]

# Split the 'genres' column into separate 'genre' columns
df[genre_cols] = df['genres'].str.split(',', expand=True)

df.head(20)
```

```python
# remove any row if genre1, genre2, or genre3 == #'Biography', 'Sci-Fi', 'Family', 'Documentary', 'History', 'Animation', 'Music', 'War', 'Sport', 'Musical', 'Western', 'Film-Noir', 'News', 'Adult', '\\N'
# First, we will create a set of genres to be removed.
genres_to_remove = {
    'Biography', 'Sci-Fi', 'Family', 'Documentary', 'History',
    'Animation', 'Music', 'War', 'Sport', 'Musical', 'Western',
    'Film-Noir', 'News', 'Adult', '\\N'
}

# Function to determine if a row should be removed
def should_remove(row):
    return any(genre in genres_to_remove for genre in row)

# Filter the DataFrame to exclude rows with genres to be removed
df_filtered = df[~df[genre_cols].apply(should_remove, axis=1)]

# df_filtered now contains only the rows where none of the genres match the genres to remove
df_filtered.reset_index(drop=True, inplace=True)

df_filtered.head(20)
```

```python
# A function to create a DataFrame with genre pairs and their counts.
def create_genre_pairs(df):
    # Split the genres string into a list of genres
    df['genres'] = df['genres'].str.split(',')
    # Get all pairs within the movies (including single genre as pair with itself)
    pairs = []
    for movie_genres in df['genres']:
        # Sort genres to treat different orderings as the same pair
        movie_genres = sorted(movie_genres)
        if len(movie_genres) == 1:  # Only one genre in the movie
            pairs.append((movie_genres[0], movie_genres[0]))
        else:
            # Add all pairs of genres from the movie
            for i in range(len(movie_genres)):
                for j in range(i, len(movie_genres)):
                    pairs.append((movie_genres[i], movie_genres[j]))
    # Convert the list of pairs to a DataFrame
    pairs_df = pd.DataFrame(pairs, columns=['genre1', 'genre2'])
    # Count the occurrence of each pair
    pairs_count = pairs_df.groupby(['genre1', 'genre2']).size().reset_index(name='count')
    return pairs_count.sort_values(by='count', ascending=False)

# Apply the function to the DataFrame
genre_pairs_df = create_genre_pairs(df_filtered)
genre_pairs_df.head(20)  # Display the top 10 genre pairs

# reset index
genre_pairs_df.reset_index(drop=True, inplace=True)

genre_pairs_df.head(20)
```

```python
import json
from itertools import product


# Create a list of unique genres
unique_genres = sorted(set(genre_pairs_df['genre1'].unique()).union(set(genre_pairs_df['genre2'].unique())))

# Create a mapping from genre to index
genre_to_index = {genre: idx for idx, genre in enumerate(unique_genres)}

# Initialize a square matrix with zeros
matrix_size = len(unique_genres)
matrix = [[0] * matrix_size for _ in range(matrix_size)]

# Populate the matrix with the counts
for _, row in genre_pairs_df.iterrows():
    index1 = genre_to_index[row['genre1']]
    index2 = genre_to_index[row['genre2']]
    count = row['count']
    matrix[index1][index2] += count
    if index1 != index2:
        matrix[index2][index1] += count  # Ensure the matrix is symmetrical


# Save the matrix and the index mapping to JSON
with open('../data/clean_data/genre_matrix.json', 'w') as f:
    json.dump({
        'matrix': matrix,
        'index': genre_to_index
    }, f)

# The file can now be used in a JavaScript application with VisX.
```

```python
# remove genre column from df_filtered
df_filtered = df_filtered.drop(columns=['genres'])

df_filtered.head(20)
```

```python
# Melt the DataFrame so each genre of each movie is in its own row
df_melted = df_filtered.melt(id_vars=['tconst', 'titleType', 'startYear'], 
                    value_vars=['genre1', 'genre2', 'genre3'],
                    var_name='genre_type', value_name='genre')

# Remove rows where genre is None
df_melted = df_melted[df_melted['genre'].notna()]

# Group by year and genre, then count the number of movies
genre_counts = df_melted.groupby(['startYear', 'genre']).size().reset_index(name='count')

# Pivot to get years as rows and genres as columns with movie counts as values
df_pivot = genre_counts.pivot(index='startYear', columns='genre', values='count').fillna(0)

# Convert genres with float counts to int (if necessary after pivot)
for col in df_pivot.columns:
    df_pivot[col] = df_pivot[col].astype(int)

# remove the last row
df_pivot = df_pivot.iloc[:-1]

df_pivot.tail(20)

# # Save to a JSON file
# df_pivot.to_json('/path/to/your/data.json', orient='index')
```
